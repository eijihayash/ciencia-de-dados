````python
# 3) seu código aqui
# Criei um variável com coluna para fazer os filtros somente das colunas que me interessa analisar se tem valor missing.
colunas = ['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO', 
    'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']
sinasc_colunas = sinasc[colunas].isna().sum()
sinasc_colunas
````

````python
# 4) Porque este código não deu certo?
print(sinasc.shape)
sinasc_filtrado = sinasc[sinasc['APGAR5']].dropna()
````

````python
# 4) apagando os valores nulos da coluna APGAR5
print(sinasc.shape)
sinasc['APGAR5'].dropna().shape
````

````python
# 5) seu código aqui
colunas = ['ESTCIVMAE', 'CONSULTAS']
for coluna in colunas:
    sinasc[sinasc[coluna].isna()] = 9
    
print(sinasc['ESTCIVMAE'].unique())
print(sinasc['CONSULTAS'].unique())
````

````python
# 6) Seu código aqui
sinasc[sinasc['QTDFILVIVO'].isna()] = 0
sinasc['QTDFILVIVO'].unique()
````

````python
# 7) seu código aqui
mediana = sinasc['QTDFILVIVO'].median()
sinasc['QTDFILVIVO'].fillna(mediana)

#substitui os valores faltantes por mediana, pois acredito que o ponto central seja mais acertivo.
````

````python
# 8) seu código aqui
sinasc.loc[(sinasc['APGAR5'] == 0) & (sinasc['APGAR5'] <= 3), 'CLASSIFICACAO_APAGAR5'] = 'asfixia severa'
sinasc.loc[(sinasc['APGAR5'] >= 4) & (sinasc['APGAR5'] <= 5), 'CLASSIFICACAO_APAGAR5'] = 'asfixia moderada'
sinasc.loc[(sinasc['APGAR5'] >= 6) & (sinasc['APGAR5'] <= 7), 'CLASSIFICACAO_APAGAR5'] = 'asfixia leve'
sinasc.loc[(sinasc['APGAR5'] >= 8) & (sinasc['APGAR5'] <=10), 'CLASSIFICACAO_APAGAR5'] = 'normal'
sinasc['CLASSIFICACAO_APAGAR5'].value_counts()
````

````python
# 9) seu código aqui
sinasc.rename(columns=lambda x:x.lower().replace(' ', '_'), inplace = True)
sinasc
#Não entendi porque não substituiu o espaço por "_"?
````

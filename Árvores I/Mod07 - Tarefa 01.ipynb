#### 2) Vamos montar um metadados

1. Crie um dataframe com os nomes de cada variável e o tipo de dados de cada variável.
2. Adicione uma coluna nesse *dataframe* chamada "qtd_categorias" e coloque nela o número de categorias correspondente de cada variável. 
    Dica: 
        1. inicie uma lista vazia
        2. faça um for ao longo dos nomes das variáveis, 
        3. conte o número de categorias dessa variável 
        4. acumule essa informação de 3. na lista que você criou em 1. 
        5. No final, essa lista pode ser atribuída à nossa variável.
3. Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}.

````python
# Crie um dataframe com os nomes de cada variável e o tipo de dados de cada variável.
# Transformando as colunas em variaveis
variaveis = df.columns
variaveis
````

````python
# Transformando os Tipos das colunas em outra variáveis
tipo_variaveis = df.dtypes
tipo_variaveis
````

````python
# Criando DataFrame com essas duas variáveis
df2 = pd.DataFrame({'variavel': variaveis,'tipo_variavel': tipo_variaveis})
df2
````

````python
df2 = df2.reset_index(drop=True)
df2
````

````python
# Adicione uma coluna nesse dataframe chamada "qtd_categorias" e coloque nela o número de categorias correspondente de cada variável.
# usei nunique para a quantidade de categoria e criei uma nova coluna com a quantidade.

qtd_categorias=[]

for coluna in df2['variavel']:
    qtd = df[coluna].nunique()
    qtd_categorias.append(qtd)

df2['qtd_categorias'] = qtd_categorias

df2.head(15)
````

````python
# Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}.
# Não entendi muito bem este passo, pois nesse DataFrame não existe variávei qualitativo.
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
#Criando dummies para as variáveis categóricos (object) mais de 2.
df_com_dummy = pd.get_dummies(df, columns=['tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia'], drop_first = True)
df_com_dummy.head(5)
````

````python
#Transformando a coluna qualitativa que possui somente duas respostas usando map
#Sexo M para 0 e F para 1
df_com_dummy.sexo = df_com_dummy.sexo.map({'F': 1, 'M':0})
````

````python
df_com_dummy.sexo.unique()
````

````python
#posse_de_veiculo N (não) para 0 e Y (sim) para 1
df_com_dummy.posse_de_veiculo = df_com_dummy.posse_de_veiculo.map({'N': 0, 'Y':1})
````

````python
df_com_dummy.posse_de_veiculo.unique()
````

````python
#posse_de_imovel N (não) para 0 e Y (sim) para 1
df_com_dummy.posse_de_imovel = df_com_dummy.posse_de_imovel.map({'N': 0, 'Y':1})
````

````python
df_com_dummy.posse_de_imovel.unique()
````

````python
#Transformando a coluna mau (booleano) para numérico 0 bom (false) e 1 mal (True)
df_com_dummy.mau = df_com_dummy.mau.astype(int)
````

````python
df_com_dummy.mau.unique()
````

````python
df_com_dummy.head()
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
tabela_cruzada = pd.crosstab(index= [df_com_dummy.possui_email, df_com_dummy.posse_de_veiculo], columns= df_com_dummy.mau, normalize='index' )
tabela_cruzada
````

````python
tabela_cruzada_1 = pd.crosstab(df_com_dummy['posse_de_veiculo'], df_com_dummy['mau'], normalize='index')
tabela_cruzada_1
````

````python
tabela_cruzada_2 =pd.crosstab(df_com_dummy['possui_email'], df_com_dummy['mau'], normalize='index')
tabela_cruzada_2
````

Acredito que seja 'posse_de_veiculo', tem mais probabilidade da pessoa estar em default.
